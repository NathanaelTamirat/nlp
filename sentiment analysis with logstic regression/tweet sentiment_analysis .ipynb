```python
# logistic regression
def sigmoid(z):
    h=1/(1+np.exp(-1*z))
    return h
```

$$z = \theta_0 x_0 + \theta_1 x_1 + \theta_2 x_2 + ... \theta_N x_N$$
$$ h(z) = \frac{1}{1+\exp^{-z}}$$
$$J(\theta) = -\frac{1}{m} \sum_{i=1}^m y^{(i)}\log (h(z(\theta)^{(i)})) + (1-y^{(i)})\log (1-h(z(\theta)^{(i)}))\tag{5} $$
$$ Loss = -1 \times \left( y^{(i)}\log (h(z(\theta)^{(i)})) + (1-y^{(i)})\log (1-h(z(\theta)^{(i)})) \right)$$
$$\nabla_{\theta_j}J(\theta) = \frac{1}{m} \sum_{i=1}^m(h^{(i)}-y^{(i)})x^{(i)}_j \tag{5}$$
$$\theta_j = \theta_j - \alpha \times \nabla_{\theta_j}J(\theta) $$


```python
def gradinet_descent(x,y,theta,alpha,num_iters):
    m=len(y)
    for i in range(0,num_iters):
        z=np.matmul(x,theta)
        h=sigmoid(z)
        J = -1/m * (np.squeeze(np.matmul(y.T , np.log(h))) + np.squeeze(np.matmul((1 - y).T , np.log(1-h))))
        theta = theta - alpha / m * np.squeeze(np.matmul(x.T,(h-y)))[:,np.newaxis]
    J = float(J)
    return J, theta
```

```python
def extract_features(tweet,freqs,process_tweet=process_tweet):
    list_word=process_tweet(tweet)
    x=np.zeros((1,3))
    x[0,0]=1 #bais
    for word in list_word:
        x[0,1]+=freqs.get((word,1),0)
        x[0,2]+=freqs.get((word,0),0)
    assert(x.shape == (1, 3))
    return x
```

```python
#train the model
X=np.zeros((len(x_train),3))
for i in range(len(x_train)):
    X[i,:]=extract_features(x_train[i],freqs)
Y=y_train
J,theta=gradinet_descent(X,Y, np.zeros((3, 1)), 1e-9, 1500)
print(J,theta)
```

```python
#  test
def predict_tweet(tweet,freqs,theta):
    x=extract_features(tweet,freqs)
    y_pred=sigmoid(np.dot(x,theta))
    return y_pred
```

```python
#example
for tweet in ['I am happy', 'I am bad', 'this movie should have been great.', 'great', 'great great', 'great great great', 'great great great great']:
    print( '%s -> %f' % (tweet, predict_tweet(tweet, freqs, theta)))    
```

```python
def test_logisitic_regression(test_x,test_y,freqs,theta,predict_tweet=predict_tweet):
    m=len(test_y)
    yhat=[]
    for tweet in test_x:
        y_pred=predict_tweet(tweet,freqs,theta)
        if y_pred>0.5:
            yhat.append(1.0)
        else:
            yhat.append(0.0)
    yhat=np.array(yhat)[:,np.newaxis]
    accuracy=np.sum(yhat==test_y)/m
    return accuracy
```

```python
#test accuracy
test_accuracy = test_logisitic_regression(test_x, test_y, freqs, theta)
test_accuracy
```
